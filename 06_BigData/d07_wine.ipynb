```python
import pandas as pd


red_df = pd.read_csv('./csv/winequality-red.csv', sep=';', header=0, engine='python')
white_df = pd.read_csv('./csv/winequality-white.csv', sep=';', header=0, engine='python')
```

```python
red_df.to_csv('./csv/winequality-red2.csv', index=False)
white_df.to_csv('./csv/winequality-white2.csv', index=False)
```

```python
red_df.insert(0, column='type', value='red')
```

```python
red_df.head()
```

```python
white_df.insert(0, column='type', value='white')
```

```python
white_df.head()
```

```python
wine = pd.concat([red_df, white_df])
wine.head()
```

```python
wine.to_csv('./csv/wine.csv', index=False)
```

```python
wine.columns = wine.columns.str.replace(' ','_')
```

```python
wine.head()
```

```python
wine.quality.value_counts()
```

```python
wine.groupby('type')['quality'].describe()
```

```python
from scipy import stats
from statsmodels.formula.api import ols,glm
```

```python
red_wine_quality= wine.loc[wine['type'] == 'red', 'quality']
white_wine_quality= wine.loc[wine['type'] == 'red', 'quality']
```

```python
stats.ttest_ind(red_wine_quality, white_wine_quality, equal_var=False)

```

```python
wine.head()
```

```python
Rformula = 'quality~fixed_acidity + volatile_acidity + citric_acid + residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates + alcohol'
regression_result = regression_result = ols(Rformula, data=wine).fit()
```

```python
sample1 = wine[wine.columns.difference(['quality', 'type'])]
sample1 = sample1[0:5]
sample1
```

```python
sample1_predict = regression_result.predict(sample1)
sample1_predict
```
